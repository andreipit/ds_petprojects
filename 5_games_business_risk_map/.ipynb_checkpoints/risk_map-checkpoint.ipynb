```python
import numpy as np
import laplotlib_8june2018 as lpl
```

```python
def draw_risk_map():
    # axis and values names
    import matplotlib.pyplot as plt
    plt.title('Карта рисков\n Зоны рисков: зеленая - неопасные, желтая - допустимые, красная - опасные')
    plt.ylabel('Вероятность появления фактора риска')
    plt.xlabel('Возможные потери')
    import pylab
    x = range(10)
    pylab.xticks(x, ['', 'минимальные','','средние','','большие',''])
    y = range(10)
    pylab.yticks(y, ['', 'маловероятно','','вероятно','','весьма вероятно',''])

    # big grid cells
    lpl.add_line_2d(p = [(2,0),(2,6)])
    lpl.add_line_2d(p = [(4,0),(4,6)])
    lpl.add_line_2d(p = [(6,0),(6,6)])
    lpl.add_line_2d(p = [(0,4),(6,4)])
    lpl.add_line_2d(p = [(0,2),(6,2)])
    lpl.add_line_2d(p = [(0,6),(6,6)])

    # paint areas
    def fill_area(x=(0,2), y=(0,2), resolution=10, pixel_radius=500, clr=(1,1,0)):
        for i in np.linspace(x[0], x[1], resolution):
            for j in np.linspace(y[0], y[1], resolution):
                lpl.add_points_2d(points = [(i,j)], anns=[''], size=pixel_radius, clr=clr)    
    greeen=(0.6,1,0.6)
    fill_area(x=(0,2), y=(0,2), clr=greeen)    
    fill_area(x=(2,4), y=(0,2), clr=greeen)    
    fill_area(x=(0,2), y=(2,4), clr=greeen)  
    red=(1,0.6,0.6)
    fill_area(x=(4,6), y=(2,4), clr=red)    
    fill_area(x=(4,6), y=(4,6), clr=red)    
    fill_area(x=(2,4), y=(4,6), clr=red)    
    yellow=(1,1,0)
    fill_area(x=(2,4), y=(2,4), clr=yellow)    
    fill_area(x=(4,6), y=(0,2), clr=yellow)    
    fill_area(x=(0,2), y=(4,6), clr=yellow)
    fill_area(x=(1,2), y=(3,4), clr=yellow)
    fill_area(x=(3,4), y=(1,2), clr=yellow)
    fill_area(x=(4,5), y=(2,3), clr=yellow)
    fill_area(x=(2,3), y=(4,5), clr=yellow)

    # zone names
    # lpl.add_points_2d(points = [(0,0)], anns=['Зона неопасных рисков'], size=1, clr=(1,0,0))    
    # lpl.add_points_2d(points = [(0,4.5)], anns=['Зона допустимых рисков'], size=1, clr=(1,0,0))    
    # lpl.add_points_2d(points = [(3.5,5.5)], anns=['Зона опасных рисков'], size=1, clr=(1,0,0))    


    lpl.add_points_2d(points = [(5.5,3.5)], anns=['Игры на 230к устаревшие (никто не купит и не сможем вернуть по франшизе)'], size=100, clr=(0,0,1), bias=(-4,0.2))    
    lpl.add_points_2d(points = [(1,1)], anns=['Поставщик не будет продавать игры со скидкой'], size=100, clr=(0,0,1))    
    lpl.add_points_2d(points = [(2,5)], anns=['Спрос на игры и антикафе будет очень низкий'], size=100, clr=(0,0,1))    
    lpl.add_points_2d(points = [(0.5,1.4)], anns=['Сотрудники откажутся работать'], size=100, clr=(0,0,1))    
    lpl.add_points_2d(points = [(2.1,2.1)], anns=['Арендодатель не сдаст нам помещение'], size=100, clr=(0,0,1))    
    lpl.add_points_2d(points = [(5,0.7)], anns=['Оштрафуют пожарники и т.п.'], size=100, clr=(0,0,1))    
    lpl.add_points_2d(points = [(4,0.1)], anns=['Нас обворуют'], size=100, clr=(0,0,1))    


    import matplotlib.pyplot as plt


    lpl.draw_vectors_2d([[0,0]], AXIS_LENGTH=6 )
    
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [12,9]
draw_risk_map()    
```
