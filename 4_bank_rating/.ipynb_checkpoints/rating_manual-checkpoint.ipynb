# optional - smart search

```python
import pandas as pd

MAX_INTERVAL = 1    # "123" vs "1aaa2aaa3" =>   interval = 3 (век  =  московский кредитный банк - must be bad)
COUNT_TOLERANCE = 1 # "123" vs "1234567890" =>  count_ratio = 1.0   ;  when 1 - all chars from short must be in long -> few words, when 0.1 - only 1 char may be -> many words
ORDER_TOLERANCE = 1 # "123" vs "1234567890" =>  ordered_ratio = 1.0 ;  percent of chars from short (that are in long), which must have right order. 0 - like no order needed/
MAX_LENGTH_DIFFERENCE = 1 #"123" vs "1234567890" => MAX_LENGTH_DIFFERENCE = 5:  "123"=="123456789";  MAX_LENGTH_DIFFERENCE = 1: "123"!="123456789" "123"=="1234"


def strInSeries(word,ser):
    word = word.strip().lower()
    max_count_ratio = 0
    max_ordered_ratio = 0
    
    #for each element in series
    for i in range(0, len(ser)):
        
        #prepare vars
        el = ser[i].strip().lower()
        count = 0   # how many chars from short word are in long
        ordered = 0 # how many chars from short word are in long and how many have right order -> [pos of char 1] < [pos of char 2]
        last_index = -1 #bookkeeping, last saved char position


        #assign short and long words
        if len(word)>len(el): 
            short = el
            long = word
        else: 
            short = word
            long = el

        # for each char in short word find count and order
        for j in range(0, len(short)):
            if short[j] in long:
                count += 1
                
                #optional: find how good is chars order;  newLetter + prevLetterPos => newIsAfter? + getNewPos (find all poses and get min value after previous char index)
                if ORDER_TOLERANCE != 0: 
                    #[pos for pos, char in enumerate('shak#spea#e') if char == '#'] => [4, 9]
                    all_char_poses = [pos for pos, char in enumerate(long) if char == short[j]] 
                    for k in range(0,len(all_char_poses)): # for each such char in long word:
                        if all_char_poses[k] > last_index: # if char is after last saved
                            if (all_char_poses[k] - last_index <= MAX_INTERVAL): # if interval is not too big (втб24  =  внешторгбанк needs max=4)
                                last_index = all_char_poses[k] # save it
                                ordered += 1
                                break
                            elif ((all_char_poses[k]==0) | (long[all_char_poses[k]-1]==' ')): # for abbreviature (if it is first letter or after space) (внеш тоsssssssssssssssssssssрг банк, even if max=0)
                                last_index = all_char_poses[k] # save it
                                ordered += 1
                                break
                # optional: find how far are char from previous
        
        # count ratios
        count_ratio = 0
        ordered_ratio = 0
        try: count_ratio = count/len(short)
        except ZeroDivisionError: 
            #print('count_ratio error in ', short)
            count_ratio = 0
        try: ordered_ratio = ordered/count
        except ZeroDivisionError: 
            #print('ordered_ratio error in ', short)
            ordered_ratio = 0
            
        # save max values
        if (count_ratio >= COUNT_TOLERANCE) & (ordered_ratio >= ORDER_TOLERANCE): #only if they are valid
            if count_ratio > max_count_ratio: max_count_ratio = count_ratio
            if ordered_ratio > max_ordered_ratio: max_ordered_ratio = ordered_ratio
                
#             print
#         if count_ratio >= COUNT_TOLERANCE: print(short, " in ", long, " has good COUNT_TOLERANCE=", count_ratio)
#         else: print(short, " in ", long, " has bad COUNT_TOLERANCE=", count_ratio)
#         if ordered_ratio >= ORDER_TOLERANCE: print(short, " in ", long, " has good ORDER_TOLERANCE=", ordered_ratio)
#         else: print(short, " in ", long, " has bad ORDER_TOLERANCE=", ordered_ratio)
#         if (count_ratio >= COUNT_TOLERANCE) & (ordered_ratio >= ORDER_TOLERANCE): print(short, " in ", long, " has GOOD count_ratio=", count_ratio, " and ordered_ratio=", ordered_ratio)
        
        if (count_ratio >= COUNT_TOLERANCE) & (ordered_ratio >= ORDER_TOLERANCE) & (len(long)-len(short) <= MAX_LENGTH_DIFFERENCE): print(short, " = ", long)
            
    
        
        
    if (max_count_ratio >= COUNT_TOLERANCE) & (max_ordered_ratio >= ORDER_TOLERANCE): #return True
        if (len(long)-len(short)) > MAX_LENGTH_DIFFERENCE: 
            #print(word, " != dead")
            return False #if length is very different - finish function
        else: 
            #print(word, " = dead")
            return True
    else: 
        #print(word, " != dead")
        return False
```

```python
import time
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import pandas as pd
import numpy as np

#https://bankiros.ru/bank/rating/top30

fhand = urllib.request.urlopen('https://bankiros.ru/bank/rating/top30')
soup = BeautifulSoup(fhand, 'html.parser');
tags = soup.findAll("a", class_="bank-name")
with open('cbr_2016_downloaded.txt', 'w', encoding='utf8') as f:
    f.write('Банк' + '\n')
    for tag in tags:
        f.write(tag.text + '\n')

cbr_2016_downloaded = pd.read_csv('cbr_2016_downloaded.txt', skiprows=0, delimiter="\t", usecols=np.arange(0, 1).tolist()).reset_index(drop=True)
cbr_2016_downloaded
```

# read all saved csv

```python
import pandas as pd
import numpy as np


forbes_2016 = pd.read_csv('forbes_2016.txt', skiprows=0, index_col=0, delimiter="\t").reset_index(drop=True)
forbes_2017 = pd.read_csv('forbes_2017.txt', skiprows=0, index_col=0, delimiter="\t").reset_index(drop=True)
forbes_2018 = pd.read_csv('forbes_2018.txt', skiprows=0, index_col=0, delimiter="\t").reset_index(drop=True)
dead_2016 = pd.read_csv('dead_2016.txt', skiprows=0, index_col=0, delimiter="\t", usecols=np.arange(0, 2).tolist()).reset_index(drop=True)
dead_2017 = pd.read_csv('dead_2017.txt', skiprows=0, index_col=0, delimiter="\t", usecols=np.arange(0, 2).tolist()).reset_index(drop=True)
dead_2018 = pd.read_csv('dead_2018.txt', skiprows=0, index_col=0, delimiter="\t", usecols=np.arange(0, 2).tolist()).reset_index(drop=True)
set_2016 = set(dead_2016['Банк'])
set_2017 = set(dead_2017['Банк'])
set_2018 = set(dead_2018['Банк'])

def mark_2016(row):
    if row['Банк'] in set_2016: return 1
    #if strInSeries(row['Банк'],dead_2016['Банк']): return 1
    else: return 0
def mark_2017(row):
    if row['Банк'] in set_2017: return 1
    #if strInSeries(row['Банк'],dead_2017['Банк']): return 1
    else: return 0
def mark_2018(row):
    if row['Банк'] in set_2018: return 1
    #if strInSeries(row['Банк'],dead_2018['Банк']): return 1
    else: return 0
    
forbes_2016['dead'] = forbes_2016.apply(mark_2016,axis=1)
forbes_2017['dead'] = forbes_2017.apply(mark_2017,axis=1)
forbes_2018['dead'] = forbes_2018.apply(mark_2018,axis=1)

print('dead in forbes: 2016 =', forbes_2016[forbes_2016['dead']==1].count()[0], ' 2017 =', forbes_2017[forbes_2017['dead']==1].count()[0],  ' 2018 =', forbes_2017[forbes_2018['dead']==1].count()[0])
```

```python
cbr_2016 = pd.read_csv('cbr_2016.txt', skiprows=0, index_col=0, delimiter="\t", usecols=np.arange(0, 2).tolist()).reset_index(drop=True)
cbr_2017 = pd.read_csv('cbr_2017.txt', skiprows=0, index_col=0, delimiter="\t", usecols=np.arange(0, 2).tolist()).reset_index(drop=True)
cbr_2018 = pd.read_csv('cbr_2018.txt', skiprows=1, delimiter="\t", usecols=np.arange(0, 1).tolist())

cbr_2016['dead'] = cbr_2016.apply(mark_2016,axis=1)
#cbr_2016_downloaded['dead'] = cbr_2016_downloaded.apply(mark_2016,axis=1)
cbr_2017['dead'] = cbr_2017.apply(mark_2017,axis=1)
cbr_2018['dead'] = cbr_2018.apply(mark_2018,axis=1)

print('dead in cbr: 2016 =', cbr_2016[cbr_2016['dead']==1].count()[0], ' 2017 =', cbr_2017[cbr_2017['dead']==1].count()[0],  ' 2018 =', cbr_2018[cbr_2018['dead']==1].count()[0])
```

# Visualize cbrf and forbes mistakes

```python
def draw_points_2d(v2=[[0,0],[0,0],[0,0]], size =[300,200,500], clr =[(1.,0.,0.), (0.,1.,0.), (0.,0.,1.)], annotation =['name1','name2','name3'], an_clr=(0,0,1), an_antiheight=3, grid_size=[-4,5], axis_length=[5,1], origin=[0,0], title="plot 01", ylabel="oy", xlabel="ox"):
    import numpy as np
    import matplotlib.pyplot as plt
    x = np.arange(grid_size[0], grid_size[1]).tolist() #grid wide
    y = np.arange(grid_size[0], grid_size[1]).tolist() #grid height
    for v in v2:
        plt.scatter([v[0]], [v[1]], s=size[v2.index(v)], c=clr[v2.index(v)]) # plt.scatter([0.1,0.5], [0.4,0.8], s=[300, 5])
        #plt.annotate(annotation[v2.index(v)], (v[0],v[1]), color=an_clr)
        plt.annotate(annotation[v2.index(v)], (v[0],v[1]), color=an_clr, xytext=(v[0]+v[0]/an_antiheight,v[1]+v[0]/an_antiheight), #textcoords='offset points', 
            ha='right', va='bottom', bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5), arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0') )
    # axis:
    plt.plot([origin[0],origin[0]], [origin[1],origin[1]+axis_length[1]]) #oy
    plt.plot([origin[0],axis_length[0]], [origin[1],origin[1]]) #ox
    plt.grid()
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.show()
    #plt.savefig('books_read.png')
    
#draw_points_2d(v2=[[1,2017],[2,2017],[3,2018]], size =[300,200,500], clr =[(1.,0.,0.), (0.,1.,0.), (0.,0.,1.)], annotation =['name1','name2','name3'], an_clr=(0,0,1), an_antiheight=5, grid_size=[-4,5], axis_length=[5,1], origin=[0,2016])
```
