```python
import pandas as pd
dfm = pd.read_csv('magnesium.csv', skiprows=0, index_col=0)
dfm.head()
```

```python
dfp = pd.read_csv('potassium.csv', skiprows=0, index_col=0)
dfp.head()
```

```python
import numpy as np
df = pd.merge(dfm, dfp, how='outer', left_index=True, right_index=True)
df['Разница потребностей, %'] = np.abs(df['Процент суточной потребности магния']-df['Процент суточной потребности калия'])
df['Порция со 100% маг, г'] = 100 * 100  / df['Процент суточной потребности магния']
df['Порция со 100% кал, г'] = 100 * 100  / df['Процент суточной потребности калия']
# df['Разница порций 100-100, %'] = np.abs(df['Порция со 100% маг, г']-df['Порция со 100% кал, г'])
# df.loc[['Отруби пшеничные', 'Арахис']]

print('Самые магниевые')
df.sort_values(by=['Порция со 100% маг, г'],ascending=True).head()
```

```python
print('Самые калиевые')
df.sort_values(by=['Порция со 100% кал, г'],ascending=True).head()
```

# Съел менее 400 г в день и все покрыл (+- 10%)

```python
limit = 400
balance = 10
print('Съел менее', limit, 'г в день и все покрыл')
# harmonic = df.sort_values(by=['Разница потребностей, %'],ascending=True).head(50)

df[(df['Разница потребностей, %']<balance) & (df['Порция со 100% кал, г']<limit) & (df['Порция со 100% маг, г']<limit)].sort_values(by=['Порция со 100% маг, г'],ascending=True)
```
