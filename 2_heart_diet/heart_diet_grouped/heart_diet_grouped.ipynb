# Load magnese

```python
import re
import pandas as pd
import numpy as np


def cleaner(row):
    # delete " мг"
    data = row[['Содержание магния в 100гр']]
    if re.search('^.*\ \м\г$', data[0]):
        new_value = re.findall(r'^(.*)\ \м\г$', data[0])
        data = data.str.replace('^.*\ \м\г$' , new_value[0].rstrip())
    row['Содержание магния в 100гр'] = data['Содержание магния в 100гр']
    
    # delete %
    data = row[['Процент суточной потребности магния']]
    if re.search('^.*\%$', data[0]):
        new_value = re.findall(r'^(.*)\%', data[0])
        data = data.str.replace('^.*\%$' , new_value[0].rstrip())
    row['Процент суточной потребности магния'] = data['Процент суточной потребности магния']
    
    # convert to numeric
    row['Содержание магния в 100гр'] = pd.to_numeric(row['Содержание магния в 100гр'], errors='coerce')
    row['Процент суточной потребности магния'] = pd.to_numeric(row['Процент суточной потребности магния'], errors='coerce')

    return row

mag_grain = pd.read_csv('mag_grain.csv', skiprows=0, index_col=0, delimiter="\t").apply(cleaner, axis=1)
mag_dried = pd.read_csv('mag_dried.csv', skiprows=0, index_col=0, delimiter="\t").apply(cleaner, axis=1)
mag_fruit = pd.read_csv('mag_fruit.csv', skiprows=0, index_col=0, delimiter="\t").apply(cleaner, axis=1)

mag_grain['Тип'] = "крупа"
mag_dried['Тип'] = "сухофрукты"
mag_fruit['Тип'] = "фрукты"

mag_GDF = mag_grain.append(mag_dried).append(mag_fruit)
# mag_GDF.head()
mag_GDF.sort_values(by=['Содержание магния в 100гр'],ascending=False).head()


# for group, frame in mag_GDF.groupby('Тип'):
#     avg = np.average(pd.to_numeric(frame['Содержание магния в 100гр'], errors='coerce'))
#     print(group, avg)

#     print(frame)
```

# Load potassium

```python
import re
import pandas as pd
import numpy as np


def cleaner(row):
    # delete " мг"
    data = row[['Содержание калия в 100гр']]
    if re.search('^.*\ \м\г$', data[0]):
        new_value = re.findall(r'^(.*)\ \м\г$', data[0])
        data = data.str.replace('^.*\ \м\г$' , new_value[0].rstrip())
    row['Содержание калия в 100гр'] = data['Содержание калия в 100гр']
    
    # delete %
    data = row[['Процент суточной потребности калия']]
    if re.search('^.*\%$', data[0]):
        new_value = re.findall(r'^(.*)\%', data[0])
        data = data.str.replace('^.*\%$' , new_value[0].rstrip())
    row['Процент суточной потребности калия'] = data['Процент суточной потребности калия']
    
    # convert to numeric
    row['Содержание калия в 100гр'] = pd.to_numeric(row['Содержание калия в 100гр'], errors='coerce')
    row['Процент суточной потребности калия'] = pd.to_numeric(row['Процент суточной потребности калия'], errors='coerce')

    return row

pot_grain = pd.read_csv('pot_grain.csv', skiprows=0, index_col=0, delimiter="\t").apply(cleaner, axis=1)
pot_dried = pd.read_csv('pot_dried.csv', skiprows=0, index_col=0, delimiter="\t").apply(cleaner, axis=1)
pot_fruit = pd.read_csv('pot_fruit.csv', skiprows=0, index_col=0, delimiter="\t").apply(cleaner, axis=1)

pot_grain['Тип'] = "крупа"
pot_dried['Тип'] = "сухофрукты"
pot_fruit['Тип'] = "фрукты"

pot_GDF = pot_grain.append(pot_dried).append(pot_fruit)
# pot_GDF.head()
pot_GDF.sort_values(by=['Содержание калия в 100гр'],ascending=False).head()
```

# Find food nearest to 65% potassium, 22% magnese in 100g

```python
target_mag = 22
target_pot = 65
grain = df.copy()
grain = grain[['Процент суточной потребности калия', 'Процент суточной потребности магния', 'Тип']]
grain['abs error'] = np.abs(target_pot - grain['Процент суточной потребности калия']) + np.abs(target_mag - grain['Процент суточной потребности магния'])
# grain = grain.loc[grain['Тип'] == 'крупа']
grain.sort_values(by=['abs error'],ascending=True).head(10)
```

# Сколько и чего добавить в отруби пшен 70 г (35%К, 78%М)?
# Find food nearest to 65% potassium, 22% magnese in ANY weight

```python
mag_abs_distance = target_mag - grain['Процент суточной потребности магния']
pot_abs_distance = target_pot - grain['Процент суточной потребности калия']
diff_dist = mag_abs_distance - pot_abs_distance
grain['scaled error'] = np.abs(diff_dist)
grain['weight with pot=65%'] = (100 / grain['Процент суточной потребности калия']) * 100
grain.sort_values(by=['scaled error'],ascending=True).head(10)
```
